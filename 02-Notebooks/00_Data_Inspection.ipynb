```python
resultat = {}
```

```python
def loadDataset(file, dico):
    """
    Loads a keyword file and transforms it into a dictionary 
    initialized to 0 for each word.

    Args:
        file (str): Path to the corresponding keyword file.
        dico (dict): The dictionary to be populated for the program.
    """
    with open(file, 'r') as file:
        for line in file:
            resultat[line.split("\n")[0]] = 0
```

```python
loadDataset("keywords", resultat)
resultat.keys()
```

```python
def recordic(file, dico):
    """
    Parses a dataset and counts the occurrences of dictionary keys.

    Args:
        file (str): Path to the dataset.
        dico (dict): Dictionary of keywords.
    """
    dico["LINE"] = 0
    with open(file, 'r') as file:
        for e in file:
            dico["LINE"] += 1
            for k in dico.keys():
                if(k in e):
                    dico[k] +=1
```

```python
recordic("MillionsTweets.csv", resultat)
```

```python
def rapportR(dico, out):
    """
    Generates a report for a dataset based on keywords that are either 
    related to or constituent of its formation.

    Args:
        dico (dict): Dictionary containing the keywords and their counts.
        out (str): Path to the output report file.
    """
    with open(out, 'a') as file:
        for k in dico.keys():
            if(k != "LINE"):
                file.write(f"{k}:{dico[k]}/{round(dico[k]/dico['LINE']*100, 4)}%\n")
```

```python
rapportR(resultat, "MillionsTweetsRapport.txt")
```
